```python
#importing libraries

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
%matplotlib inline
digits = load_digits()
```

# DATA SET 1

```python
#detertmining the total number of images and labels
print("image data shape", digits.data.shape)
print("label data shape", digits.target.shape)
```

```python
#displaying some of the images 
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' %label, fontsize=20)
```

```python
#dividing data into train and test data
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.23,random_state=2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#importing logistic regression model
from sklearn.linear_model import LogisticRegression
```

```python
#Making instance of  model and training it
logisticreg = LogisticRegression()
logisticreg.fit(X_train,y_train)
```

```python
#Predicting output of the first element of the test set
print(logisticreg.predict(X_test[0].reshape(1,-1)))
```

```python
#predicting the output of the first 10 elements of the test set
logisticreg.predict(X_test[0:10])
```

```python
#Predicting for the entire data set
predictions = logisticreg.predict(X_test)
```

```python
#Determining the accuracy of the model
score = logisticreg.score(X_test,y_test)
print(score)
```

```python
#Respresenting confusion matrix in heat map
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression(random_state= 0)
logit.fit(X_train, y_train) 
y_predicted = logit.predict(X_test)

```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f",linewidths=.5,square=True, cmap='Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title='Accuracy Score : 0.9420289855072463'.format(score)
plt.title(all_sample_title, size = 15)
```
